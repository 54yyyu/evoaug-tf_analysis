```python
import os
import numpy as np
from six.moves import cPickle
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
output_dir = '../results_downsample/'
expt_name = 'deepstarr_ITD_'
downsamples = np.array([25000, 100000, 200000, 300000, 402278])
task = 0


aug = []
finetune = []
supervised = []
for downsample in downsamples:
    with open(os.path.join(output_dir, expt_name+str(downsample)+'_results.pickle'), 'rb') as f:
        results_aug = cPickle.load(f)
        results_finetune = cPickle.load(f)
        results_standard = cPickle.load(f)

    scores = []
    for results in results_aug:
        scores.append(results[0][task])
    scores = np.array(scores)
    aug.append(scores)
    
    scores = []
    for results in results_finetune:
        scores.append(results[0][task])
    scores = np.array(scores)
    finetune.append(scores)
    
    scores = []
    for results in results_standard:
        scores.append(results[0][task])
    scores = np.array(scores)
    supervised.append(scores)
    
    
```

```python

N = len(downsamples)
fontsize=12


cmap = cm.ScalarMappable(cmap='tab10')
rgb = cmap.to_rgba(np.linspace(0, 1, 8))

fig = plt.figure(figsize=(5,3))
plt.scatter(downsamples, np.mean(np.array(aug), axis=1), color=rgb[2,:], s=50)
plt.scatter(downsamples, np.mean(np.array(finetune), axis=1), color=rgb[1,:], s=50)
plt.scatter(downsamples, np.mean(np.array(supervised), axis=1), color=rgb[0,:], s=50)


mean_vals = []
std_vals = []
for i in range(len(aug)):
  mean_vals.append(np.mean(aug[i]))
  std_vals.append(np.std(aug[i]))
mean_vals = np.array(mean_vals)
std_vals = np.array(std_vals)
plt.plot(downsamples, mean_vals, color=rgb[2,:], alpha=1)
plt.fill_between(downsamples, mean_vals-std_vals, mean_vals+std_vals,
    alpha=0.3, edgecolor=rgb[2,:], facecolor=rgb[2,:])
plt.scatter(downsamples, mean_vals, color=rgb[2,:], s=50)

mean_vals = []
std_vals = []
for i in range(len(finetune)):
  mean_vals.append(np.mean(finetune[i]))
  std_vals.append(np.std(finetune[i]))
mean_vals = np.array(mean_vals)
std_vals = np.array(std_vals)
plt.plot(downsamples, mean_vals, color=rgb[1,:], alpha=1)
plt.fill_between(downsamples, mean_vals-std_vals, mean_vals+std_vals,
    alpha=0.3, edgecolor=rgb[1,:], facecolor=rgb[1,:])
plt.scatter(downsamples, mean_vals, color=rgb[1,:], s=50)

mean_vals = []
std_vals = []
for i in range(len(supervised)):
  mean_vals.append(np.mean(supervised[i]))
  std_vals.append(np.std(supervised[i]))
mean_vals = np.array(mean_vals)
std_vals = np.array(std_vals)
plt.plot(downsamples, mean_vals, color=rgb[0,:], alpha=1)
plt.fill_between(downsamples, mean_vals-std_vals, mean_vals+std_vals,
    alpha=0.3, edgecolor=rgb[0,:], facecolor=rgb[0,:])
plt.scatter(downsamples, mean_vals, color=rgb[0,:], s=50)


ax = plt.gca();
ax.legend(['Augment', 'Finetune', 'Standard'],
          loc='lower right', fontsize=fontsize, frameon=False)

plt.setp(ax.get_yticklabels(), fontsize=fontsize)
plt.setp(ax.get_xticklabels(), fontsize=fontsize)
plt.ylabel("Pearson's r", fontsize=fontsize);
plt.xlabel('Training Set size', fontsize=fontsize);
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xticks([0, 100000, 200000, 300000, 400000])
fig.savefig('../results_downsample/downsample_dev.pdf', format='pdf', dpi=300, bbox_inches='tight')
```
