```python
import os
import numpy as np
from six.moves import cPickle
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
output_dir = '../result/'
expt_name = 'deepstarr_'
task = 0

augs = ['R', 'N', 'M', 'T', 'D', 'I',  'ITD', 'RITDMN']
aug_names = ['RC', 'Noise', 'Mutation', 'Translocation', 'Deletion', 'Insertion', 'Ins+Del+Tra', 'All']


aug = []
finetune = []
for aug_name in augs:
    with open(os.path.join(output_dir, expt_name+aug_name+'_results.pickle'), 'rb') as f:
        results_aug = cPickle.load(f)
        results_finetune = cPickle.load(f)

    scores = []
    for results in results_aug:
        scores.append(results[0][task])
    scores = np.array(scores)
    aug.append(scores)
    
    scores = []
    for results in results_finetune:
        scores.append(results[0][task])
    scores = np.array(scores)
    finetune.append(scores)
    
    
```

```python
output_dir = 'drive/MyDrive/augmentation/'


N = len(aug_names)
fontsize=12


cmap = cm.ScalarMappable(cmap='tab10')
rgb = cmap.to_rgba(np.linspace(0, 1, 8))  

fig = plt.figure(figsize=(6,3))

bplot3 = plt.boxplot(aug, 
                    notch=False, 
                    patch_artist=True, 
                    widths=0.22, 
                    positions=np.array(range(N))-.15, 
                    medianprops=dict(color="black",linewidth=1), 
                    showfliers=True);
for patch in bplot3['boxes']:
  patch.set_facecolor(rgb[2,:])
  patch.set_edgecolor('black')


bplot1 = plt.boxplot(finetune, 
                    notch=False, 
                    patch_artist=True, 
                    widths=0.22, 
                    positions=np.array(range(N))+.15, 
                    medianprops=dict(color="black",linewidth=1), 
                    showfliers=True);
for patch in bplot1['boxes']:
  patch.set_facecolor(rgb[1,:])
  patch.set_edgecolor('black')


ax = plt.gca();
ax.legend([bplot3["boxes"][0], bplot1["boxes"][0]], ['Augment', 'Finetune'], 
          loc='upper left', fontsize=fontsize, frameon=False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.setp(ax.get_yticklabels(), fontsize=fontsize)
plt.setp(ax.get_xticklabels(), fontsize=fontsize)
plt.ylabel("Pearson's r", fontsize=fontsize);
plt.xticks(range(N), aug_names, fontsize=fontsize, rotation=45, horizontalalignment="right");
plt.plot([-.5, 7.5], [.672, .672], '--k')
fig.savefig('../result/deepstarr_dev.pdf', format='pdf', dpi=200, bbox_inches='tight')
```
